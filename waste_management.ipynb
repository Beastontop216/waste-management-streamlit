```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("Waste_Management_Dataset_200_Rows.csv")

# Average recycling rate by region
avg_recycling_by_region = df.groupby("Region")["Recycling_Rate (%)"].mean()

# Plotting
plt.figure(figsize=(10, 6))
avg_recycling_by_region.plot(kind='bar', color='green')
plt.title("Average Recycling Rate by Region", fontsize=16)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Recycling Rate (%)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```

```python
# Aggregate total waste generated per year
total_waste_per_year = df.groupby("Year")["Total_Waste_Generated (tons)"].sum()

# Plotting the total waste generated over time
plt.figure(figsize=(10, 6))
total_waste_per_year.plot(kind='line', marker='o', color='blue')
plt.title("Total Waste Generated Over Time", fontsize=16)
plt.xlabel("Year", fontsize=12)
plt.ylabel("Total Waste Generated (tons)", fontsize=12)
plt.grid(True)
plt.show()
```

```python
# Aggregate average percentages of disposal methods by region
disposal_methods = df.groupby("Region")[["Recycling_Rate (%)", "Landfill_Usage (%)", 
                                         "Incineration (%)", "Composting (%)"]].mean()

# Plotting stacked bar chart
disposal_methods.plot(kind='bar', stacked=True, figsize=(12, 7), colormap="Set3")
plt.title("Waste Disposal Methods by Region", fontsize=16)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Percentage (%)", fontsize=12)
plt.legend(title="Disposal Methods", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

# Boxplot to show the distribution of waste generated by region
plt.figure(figsize=(10, 6))
sns.boxplot(x="Region", y="Total_Waste_Generated (tons)", data=df)
plt.title("Distribution of Waste Generation by Region", fontsize=16)
plt.xlabel("Region", fontsize=12)
plt.ylabel("Total Waste Generated (tons)", fontsize=12)
plt.xticks(rotation=45)
plt.show()
```
